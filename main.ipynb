```python
from openai import OpenAI
import os
from dotenv import load_dotenv
import pandas as pd
import numpy as np
import urllib.request

load_dotenv()

OPENAI_API_KEY = os.getenv('openai_token')

client = OpenAI(
    api_key=OPENAI_API_KEY,
)

df = pd.read_csv('Housing.csv')

context = "Image you are a real estate manager who wants to sell a house via a post on social media. You need to show the advantages of the house and develope an idea of a good deal."

example = ""

houses = list()

for i in range(20):
    index = np.random.randint(0,df.shape[0], 1)[0]
    house_prop = ""
    for i, j in df.iloc[index].items():
        house_prop += f"{i}={j}\n"
        
    houses.append(house_prop)
        
    content = f"Write a short post for an Instagram real estate account. The house properties are: {house_prop}"
    
    completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
        {"role": "system", "content": context},
        {"role": "user", "content": content}
      ]
    )
    
    example += house_prop
    example += "// all parameters used to generate this sample\n"
    example += completion.choices[0].message.content + "\n"
    example += "------------------------ //\n"
    
with open('EXAMPLES.txt', 'w', encoding="utf-8") as f:
    f.writelines(example)
```

```python
print(example)
```

```python
images = list()

for i in houses:
    response = client.images.generate(
      model="dall-e-3",
      prompt=f"I NEED to test how the tool works with extremely simple prompts. DO NOT add any detail, just use it AS-IS: No text on the image. Generate a realistic inside photo of a typical American house with the following properties: {i}",
      size="1024x1024",
      quality="standard",
      n=1,
    )
    
    images.append(response.data[0].url)
```

```python
for i in images:
    print(i)
```
